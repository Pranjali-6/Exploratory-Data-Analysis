## Avg rate of hotels not cancelled

```python
data = data[data['adr']>0]
```

```python
data['avg_adr'] = data['adr']/(data['adults']+data['children'])
```

```python
resort = data[(data['hotel'] == 'Resort Hotel') & (data['is_canceled'] == 0)]
city = data[(data['hotel'] == 'City Hotel') & (data['is_canceled'] == 0)]
```

```python
resort_hotel = resort.groupby(['arrival_date_month'])['avg_adr'].mean().reset_index()
#resort_hotel
```

```python
city_hotel = city.groupby(['arrival_date_month'])['avg_adr'].mean().reset_index()
#city_hotel
```

```python
final = resort_hotel.merge(city_hotel,on = 'arrival_date_month')
final.columns = ['month','price_for_resort','price_for_city_hotel']
final
```

```python
sort_order=['January','February','March','April','May','June','July','August','September','October','November','December']
final.index=pd.CategoricalIndex(final['month'],categories=sort_order,ordered=True)
final=final.sort_index().reset_index(drop=True)
final
```

```python
plt.figure(figsize=(10,10))
plt.plot(final['month'],final['price_for_city_hotel'])
plt.plot(final['month'],final['price_for_resort'])
plt.xlabel('Hotel')
plt.ylabel('Avg_adr')
plt.rcParams.update({'font.size': 50})
plt.legend(['City Hotel','Resort Hotel'])
plt.show()
```

ADR for resort hotel is very high in the months of june, july, august, september.
Duing this time city hotel prices are low.
